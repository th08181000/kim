
예측값이 NaN으로 나오는 이유는 다양한 원인이 있을 수 있습니다. 주로 다음과 같은 상황에서 발생할 수 있습니다:

모델이 불안정하거나 잘 학습되지 않았을 경우: 모델이 충분한 학습을 하지 못하거나 특정 패턴을 파악하지 못할 때, 예측값이 불안정하게 나올 수 있습니다. 이 경우 모델의 구조나 하이퍼파라미터 설정 등을 조정하여 성능을 향상시킬 수 있습니다.

입력 데이터에 결측치가 있을 경우: 모델이 학습할 때 사용된 데이터와 예측할 때 사용되는 데이터 사이에 결측치가 있을 경우, 예측값도 결측치(NaN)로 나올 수 있습니다. 이 경우 결측치 처리를 위한 적절한 전처리 과정을 추가하여 문제를 해결할 수 있습니다.

특정 특성 범위 밖의 데이터가 주어졌을 경우: 모델이 학습할 때 사용된 데이터의 특성 범위 밖의 값을 예측할 때는 정상적인 예측값이 나오지 않을 수 있습니다. 이 경우 입력 데이터의 범위를 모델이 학습한 범위로 제한하거나, 이상치 처리 등을 통해 문제를 해결할 수 있습니다.

따라서 NaN 예측값이 나오는 경우, 모델이 학습할 때 사용된 데이터와 예측할 때 사용되는 데이터, 그리고 모델 자체의 설정 등을 확인하여 원인을 파악하고 적절한 대응을 취해야 합니다.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split


lr_clf = LogisticRegression()
knn_clf = KNeighborsClassifier()

X = df.drop('총액', axis =1)
Y = df['총액']

vo_clf = VotingClassifier(estimators=[('LR', lr_clf), ('KNN', knn_clf)], voting='soft')

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=156)

vo_clf.fit(X_train, y_train)
pred = vo_clf.predict(X_test)
print('Voting 분류기 정확도: {0:.4f}'.format(accuracy_score(y_test, pred)))

classifiers = [lr_clf, knn_clf]
for classifier in classifiers:
    classifier.fit(X_train, y_train)
    pred = classifier.predict(X_test)
    class_name = classifier.__class__.__name__
    print('{0} 정확도: {1:.4f}'.format(class_name, accuracy_score(y_test, pred)))
```
